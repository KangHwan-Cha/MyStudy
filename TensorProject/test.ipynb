```python
# val_loss 소수점 자리수 4자리로 맞춤
path = 'tensorflow_cert'
for dirs, folders, files in os.walk(path):
    for f in files:
        if 'val_loss' in f:
            f_loss = re.sub('[^\.0-9]', '', os.path.splitext(f)[0])
            f_loss_change = f'{round(float(f_loss), 4):.4f}'
            pre_f = f
            aft_f = re.sub(f_loss, f_loss_change, f)
            os.rename(os.path.join(dirs, pre_f), os.path.join(dirs, aft_f))
            print(pre_f, aft_f, sep=' || ')
            # print(f_loss, f_loss_change)
            # print(f)
            # print(os.path.splitext(f))
            # print(os.path.join(dirs, f))
```

```python
import re
def tweak(filename: str, contents: str) -> str:
    url = 'https://colab.research.google.com/github/KangHwan-Cha/MyStudy/blob/main/TensorProject/'
    repattern = re.compile(r'Category\d[A-Z].*ipynb')
    check = re.search(repattern, contents)
    try:
        st, ed = check.regs[0][0], check.regs[0][1]
    except:
        return False
    ch_con = re.sub(contents[st:ed], filename, contents)
    if 'Category2A' in filename:
        print(ch_con)
    # print(check)

    return True


for file in os.listdir():
    if '.ipynb' in file and 'Category' in file:
        with open(file, 'r', encoding='utf-8') as f:
            filename = file
            con = f.readline()
            # print(con)
            tweak(filename, con)
```
